```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'ID' is the column containing image IDs
# 'features' is a DataFrame containing the features you want to use for clustering
data = pd.read_csv(r"C:\Users\Usuario\Downloads\INFO VISUALISATION\celeba_buffalo_l.csv")
image_ids = data.iloc[:, 0]  # Assuming the non-numeric column is in the first column
features = data.iloc[:, 1:552]  # Exclude the first column, which contains non-numeric data

# Choose the number of clusters (you may need to experiment with this)
num_clusters = 3

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['cluster'] = kmeans.fit_predict(features)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=3)
reduced_features = pca.fit_transform(features)

# Add the reduced features to the dataframe
data['PCA1'] = reduced_features[:, 0]
data['PCA2'] = reduced_features[:, 1]
data['PCA3'] = reduced_features[:, 2]

# Visualize the clusters in the reduced space
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for cluster in range(num_clusters):
    cluster_data = data[data['cluster'] == cluster]
    ax.scatter(cluster_data['PCA1'], cluster_data['PCA2'], cluster_data['PCA3'], label=f'Cluster {cluster}')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.legend()
ax.set_title('Clustering of Images (PCA)')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'ID' is the column containing image IDs
# 'features' is a DataFrame containing the features you want to use for clustering
data = pd.read_csv(r"C:\Users\Usuario\Downloads\INFO VISUALISATION\celeba_buffalo_l.csv")
image_ids = data.iloc[:, 0]  # Assuming the non-numeric column is in the first column
features = data.iloc[:, 1:]  # Exclude the first column, which contains non-numeric data

# Choose the number of clusters (you may need to experiment with this)
num_clusters = 3

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
data['cluster'] = kmeans.fit_predict(features)

# Apply PCA for dimensionality reduction
num_components = 5  # Adjust the number of components as needed
pca = PCA(n_components=num_components)
reduced_features = pca.fit_transform(features)

# Add the reduced features to the dataframe
for i in range(num_components):
    data[f'PCA{i + 1}'] = reduced_features[:, i]

# Pairwise scatter plots for the first five principal components
sns.pairplot(data, hue='cluster', vars=[f'PCA{i + 1}' for i in range(num_components)])
plt.suptitle('Pairwise Scatter Plots of Principal Components')
plt.show()
```
